```python
import requests

import geopandas as gpd
from shapely.geometry import shape

api_url = "https://api.energyandcleanair.org/stations?country=GB,US,TR,PH,IN,TH&format=geojson"

# Fetch data from API
response = requests.get(api_url)

# Check if the request was successful
if response.status_code == 200:
    geojson_data = response.json()
    print("Data fetched successfully!")
else:
    print(f"Failed to fetch data. HTTP Status Code: {response.status_code}")
    exit()

stations_gdf = gpd.GeoDataFrame.from_features(geojson_data["features"])
```

```python
import requests

import geopandas as gpd
from shapely.geometry import shape

api_url_country = "https://r2.datahub.io/clvyjaryy0000la0cxieg4o8o/main/raw/data/countries.geojson"

# Fetch data from API
response = requests.get(api_url_country)

# Check if the request was successful
if response.status_code == 200:
    geojson_data_country = response.json()
    print("Data fetched successfully!")
else:
    print(f"Failed to fetch data. HTTP Status Code: {response.status_code}")
    exit()

countries_gdf = gpd.GeoDataFrame.from_features(geojson_data_country["features"])
```

```python
countries_gdf.head()
```

```python
stations_gdf.columns
```

```python
countries_gdf.columns
```

```python
stations_gdf = stations_gdf.to_crs(countries_gdf.crs)
```

```python
print("Stations CRS:", stations_gdf.crs)
print("Countries CRS:", countries_gdf.crs)
```

```python
# Setting CRS as its missing
if stations_gdf.crs is None:
    stations_gdf.set_crs("EPSG:4326", inplace=True)  # Assuming WGS 84

if countries_gdf.crs is None:
    countries_gdf.set_crs("EPSG:4326", inplace=True)  # Assuming WGS 84
```

```python
stations_gdf
```

```python
stations_gdf = stations_gdf.to_crs("EPSG:3857")
countries_gdf = countries_gdf.to_crs("EPSG:3857")
```

```python
countries_gdf["area_km2"] = countries_gdf["geometry"].area / 1e6
```

```python
# Filter for countries of interest
countries_of_interest = ["United States of America", "United Kingdom", "Turkey", "Thailand", "Philippines", "India"]
filtered_countries = countries_gdf[countries_gdf["ADMIN"].isin(countries_of_interest)]
```

```python
filtered_countries
```

```python
stations_within_countries = gpd.sjoin(stations_gdf, filtered_countries, how="inner", predicate="within")
```

```python
stations_within_countries
```

```python
stations_within_countries
```

```python
stations_count = stations_within_countries.groupby("ADMIN").size().reset_index(name="num_stations")
```

```python
stations_count
```

```python
# Merge station counts with country areas
result = filtered_countries.merge(stations_count, left_on="ADMIN", right_on="ADMIN", how="left")
result["num_stations"] = result["num_stations"].fillna(0).astype(int)
```

```python
result
```

```python
# Calculate density
result["density_per_1000_sqkm"] = (result["num_stations"] / result["area_km2"]) * 1000
```

```python
result
```

```python
# Final table
final_table = result[["ADMIN", "num_stations", "area_km2", "density_per_1000_sqkm"]].rename(
    columns={
        "ADMIN": "Country Name",
        "num_stations": "Number of PM10 Stations",
        "area_km2": "Area (sq. km)",
        "density_per_1000_sqkm": "Density (stations/1000 sq. km)"
    }
)
```

```python
# Sort and display
final_table = final_table.sort_values(by="Density (stations/1000 sq. km)", ascending=False)
print(final_table.to_string(index=False))
```
